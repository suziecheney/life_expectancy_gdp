# Life Expectancy and GDP Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('all_data.csv')
df = df.rename(columns={'Country': 'country', 'Year': 'year', 'Life expectancy at birth (years)': 'life_expectancy', 'GDP': 'gdp'})
print(df.head(15))
df['country'] = df['country'].str.replace('United States of America','USA')

unique_countries= df.country.unique()
unique_years= df.year.unique()



print(unique_countries)
print(unique_years)
```

```python
perc=0.75
df_sub=df.sample(n=int(df.shape[0]*perc))


sns.scatterplot(x=df['year'], y=df['life_expectancy'], hue=df['country'], data=df)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy by Year')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.clf()
```

```python
sns.scatterplot(x=df['year'], y=df['gdp'], hue=df['country'], data=df)
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP by Year')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.clf()
```
